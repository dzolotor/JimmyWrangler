In this notebook, data sets are cleaned, transformed, and merged into a single dataset <br>
We will compare the U.S. Treasury Yield (Fed Rate) with the Trade Weighted U.S. Dollar Index, how the fed can effect the value of the dollar.<br><br>
First, we import our packages <br>

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
```

We read our first data set into a data frame <br>
This data set is the Trade Weighted U.S. Dollar Index <br>
We convert the DATE column to datetime so we can merge data sets with different date formats <br>

```python
dfForex = pd.read_csv('./DataSets/TWEXB.csv')
dfForex['DATE'] = pd.to_datetime(dfForex['DATE'])
dfForex.head()
```

Now we read our second data set into a data frame <br>
This is the Market Yield for U.S. Treasuries

```python
dfInterest = pd.read_csv('./DataSets/FRB_H15.csv')
dfInterest
```

We can drop the data we don't use <br>
<br>
It is standard to report 1-year maturities in Percent/Year, so we drop column 0 <br>
There is no multiplier, so we drop column 1 <br>
The U.S. issues their debt in their own currency and we only care about yield as a percent anyway, so we can drop column 2 <br>
We aren't concerned with the Fed's proprietary identifiers, so we will drop rows 3 and 4 <br>
<br>
We want to see how interest rates might effect the value of the dollar<br>
This will be easiest to see when we focus on only one maturity rate, so <br>
we will look at 1-year maturity notes, and all other maturity columns

```python
dfInterest = dfInterest.drop([0, 1, 2, 3, 4])
dfInterest = dfInterest.drop(dfInterest.columns[[1,2,3,5,6,7,8,9,10,11]], axis=1)
dfInterest.head()
```

Let's also rename the columns to make the more readable<br>
Again, we will format the date as a datetime so we can merge the data sets<br>
The first data set only goes back to 1995, so we can drop all rows before then<br>

```python
dfInterest = dfInterest.rename(columns={'Series Description': 'DATE'})
dfInterest.rename(columns = {list(dfInterest)[1]: 'YIELD'}, inplace = True)
dfInterest['DATE'] = pd.to_datetime(dfInterest['DATE'])
dfInterest = dfInterest[dfInterest['DATE'].dt.year >= 1996]

dfInterest.head()
```

```python
dfMerged = pd.merge(
    dfInterest,
    dfForex,
    left_on=['DATE'],
    right_on=['DATE']
)
dfMerged
```

Now we plot YIELD vs the USD Index to see how interest rate effects the value of the dollar

```python
dfMerged = dfMerged[dfMerged['YIELD'] != 'ND']
dfMerged['YIELD'] = pd.to_numeric(dfMerged['YIELD'])

dfMerged.plot(kind='scatter',x='YIELD',y='TWEXB')
plt.show()
```

We can see that as interest rates fall below 1%, the value of the dollar tends to fall <br>
Similarly, as interest rates rise above 5%, the value of the dollar tends to fall off as well <br>
<br>
However, as the Federal Reserve Bank changes economic policy (and the Chairman changes), markets may interpret their actions differently<br>
Let's break this data down into plots that only cover 5 years and see how the trend holds

```python
for year in range(1995, 2020, 5):
    df = dfMerged[dfMerged['DATE'].dt.year >= year]
    df = df[df['DATE'].dt.year < year+5]
    df.plot(kind='scatter', x='YIELD', y='TWEXB')
    
    plt.title( str(year) + '-' + str(year+5) )
    plt.ylabel('Trade Weighted U.S. Dollar Index')
    plt.xlabel('1 Year U.S. Treasury Rate')
    plt.show()
```

The trend generally holds for each of the 5 year periods<br>
